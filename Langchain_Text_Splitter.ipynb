# Character Text Splitter

- **Basic Functionality:**

- This splitter divides the text into chunks based solely on a character limit , regardless of natural text boundaries like swentences or paragraphs

- **Use Case** : it is siuitable for simple scenarios where splitting the text by a fixed number of character is sufficients , and you don't care about aplitting within words or sentences.

- **KeyParameter**:

  - chunk_size : The maximun number of characters in each chunk

  - chunk_overlap :The number of character that should overlap between consecutive chunks.

- **How it works :** It will take the input text and split it into parts by counting characters .

```python
from langchain_text_splitters import CharacterTextSplitter
```

```python
text_splitter=CharacterTextSplitter(chunk_size=2000,chunk_overlap=100)
```

```python
docs = text_splitter.split_documents(documents)
len(docs)
```

```python
sum([len(x.page_content) for x in docs])
```

```python
[len(i.page_content) for i in docs]
```

```python
text_splitter=CharacterTextSplitter(chunk_size=2000,chunk_overlap=100,separator=' ')
docs=text_splitter.split_documents(documents)
len(docs)
```
